```python
import pandas as pd
import os
from tqdm import tqdm
import string
```

```python
#world_country_and_usa_states_latitude_and_longitude_values.csv
#worldcities.csv

country_states_file = input("File path for pre-RL dataset: ")
country_states_dataset = pd.read_csv(country_states_file)

cities_file = input("File path for post-RL dataset: ")
cities_dataset = pd.read_csv(cities_file)
```

```python
country_states_dataset.head()
```

```python
cities_dataset.head()
```

```python
# disect country and US states, rename. 
country_dataset = country_states_dataset[["country", "latitude", "longitude"]]
country_dataset.rename(columns={'country': 'name'}, inplace=True)
```

```python
state_dataset = country_states_dataset[["usa_state", "usa_state_latitude", "usa_state_longitude"]]
state_dataset.rename(columns={'usa_state': 'name', 'usa_state_latitude': 'latitude', 'usa_state_longitude':'longitude'}, inplace=True)
```

```python
# dissect and standardize city dataset.
cities_dataset = cities_dataset[["city_ascii","lat","lng"]]
cities_dataset.rename(columns={'city_ascii': 'name', 'lat': 'latitude', 'lng':'longitude'}, inplace=True)
```

```python
# union all three datasets into 1. 
CCS_dataset = pd.concat([country_dataset, state_dataset, cities_dataset])
CCS_dataset.drop_duplicates(subset='name')
```

```python
CCS_dataset = CCS_dataset.dropna()
```

```python
def preprocess_name(name):
    # Remove punctuation and convert to lowercase
    return name.translate(str.maketrans('', '', string.punctuation)).lower()
```

```python
CCS_dataset['name'] = CCS_dataset['name'].apply(preprocess_name)
```

```python
CCS_dataset.to_csv('cleaned_CCS_dataset.csv', index=False)
```
